```python
#1. Расчет конверсии перехода студентов из одного модуля в другой на каждом курсе
#Создаем отдельный DataFrame и оставляем только начатые домашние работы
data_start = df[df['is_homework']==True][['module_number', 'lesson_number', 'title', 'student_id']]
data_start = data_start.groupby(['title', 'module_number', 'student_id']).agg({'lesson_number': 'min'}).reset_index()
#сгруппируем по курсам и модулям и рассчитаем количество студентов приступивших в новой домашней работе
data_start = data_start.groupby(['title', 'module_number']).count().reset_index()
#Создаем отдельный DataFrame и оставляем только выполненные домашние работы
data_done = df[(df['is_homework']==True) & (df['status'] == 'done')][['module_number', 'lesson_number', 'title', 'student_id']]
data_done = data_done.groupby(['title', 'module_number', 'student_id']).agg({'lesson_number': 'max'}).reset_index()
#сгруппируем по курсам и модулям и рассчитаем количество студентов выполневших домашние работы
data_done = data_done.groupby(['title', 'module_number']).count().reset_index()
#соединяем в один dataframe
data_start['done_works'] = data_done['student_id']
#сдвигаем на один модуль выполненные дз
shift=[]
for i_title in data['title'].drop_duplicates():
      shift.extend(data_start[data_start['title'] == i_title].iloc[:, -1].shift(1, fill_value = 0))
data_start['done_works'] = shift
#считаем конверсию
data_start['conversion'] = data_start['student_id'] / data_start['done_works']
data_start = data_start[data_start['done_works'] != 0]
data_start.head()
```
